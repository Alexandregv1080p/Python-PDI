```python
nome = input('Digite seu nome: ') #Para lermos um valor do console, atribuimos um input para uma variável
print(nome) #Para lermos damos um print
frase1 = "Olá, {}. Seja bem-vindo".format(nome) #Aqui interpolamos a variável a uma str
frase2 = f"Olá, {nome}. Seja bem-vindo" #Para utilizar a variávei dentro de uma sentença, colocamos o "f" no antes e passamos a variável dentro das chaves
print(frase1)
print(frase2)
```
