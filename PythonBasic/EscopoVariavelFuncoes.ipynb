```python
#Escopo de variáveis e função

def imprime_texto(): #Uma função se inicia com dev, seguido pelo nome atribuido e ():
  print("Texto legal!")
imprime_texto()

#Variáveis pode ser definidas de nível global e local

numero_global = 15 #Variável global
def imprime_numero():
  numero_global = 10 #Variavel local
  print(numero_global) #O 10 é atribuido para esta variavel local
imprime_numero()
print(numero_global) #O valor 15 é mantido

numero_global1 = 15
def imprime_numero():
  global numero_global1 #Aqui forçamos a variável global
  numero_global1 = 10 #Alteramos a variável em estado global
  print(numero_global1)
imprime_numero()
print(numero_global1) 
```
