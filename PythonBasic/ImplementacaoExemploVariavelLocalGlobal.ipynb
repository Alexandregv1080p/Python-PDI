```python
#Implementação de exemplo com o uso de variáveis local e global

cliente = "João"
passos = 0
status = "Aguardando pedido"

def entregar():
  global passos
  passos += 1
  status = "Pedido entregue"
  print(f"{cliente}, passos dado: {passos}, status: {status}")


def preparar_pedido():
  global passos
  passos += 1
  status = "Pedido sendo preparado"
  print(f"Passos: {passos}, status: {status}")
  entregar()

def atender():
  global passos
  passos += 1
  status = "Pedido recebido!"
  print(f"Passos: {passos}, status: {status}")
  preparar_pedido()
atender()

```
