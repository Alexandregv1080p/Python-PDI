```python
#Estruturas condicionais

nota1 = float(input(f"Digite sua 1a nota: "))
nota2 = float(input(f"Digite sua 2a nota: "))
nota3 = float(input(f"Digite sua 3a nota: "))

media = (nota1 + nota2 + nota3)/3
if media < 4:
  print("Reprovado")
elif media < 6:
  print("Recuperação")
else:
  print("Aprovado")
```
