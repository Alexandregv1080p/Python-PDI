```python
#Trabalhando com loops e estrutura aninhadas

dados = {"Rex": ["Biscoito", "Ração", "Água"], "Fifi":["Ração", "Água"]}
dogs = list(dados.keys()) #Pegamos todas as chaves contidas em dados. Ex: Rex e Fifi

j = 0

while j < len(dogs):
  cachorro = dogs[j]
  print(f"A última refeição do {cachorro}, foi: ")
  i = 0
  while i < len(dados[cachorro]):
    print(dados[cachorro][i])
    i += 1
  j += 1
```
